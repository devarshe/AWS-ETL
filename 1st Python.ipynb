```python
dog_person = True
cat_person = False
age = int(input("Enter age"))

if age < 18:
    print("You're under 18 — please get your parents' permission before choosing a dog.")
else:
    if dog_person and cat_person:
        print("We recommend a Golden Retriever — friendly and great with other pets!")
    elif dog_person and not cat_person:
        print("We recommend a Scottish Deerhound — perfect for a cat-free home.")
    elif cat_person and not dog_person:
        print("You might be barking up the wrong tree looking for a dog!")
    else:
        print("You may want to think about whether a pet is the right choice for you.")
```

```python
def calculate_total_cost(item_price, shipping_cost=50):
    total = item_price + shipping_cost
    print(f"Total cost: ₹{total}")

calculate_total_cost(1000)

calculate_total_cost(1000, 150)
```
